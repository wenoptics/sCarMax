```python
import sys
sys.path.insert(0, "D:\\Projects\\JobSeekingHelper\\venv\\Lib\\site-packages")

from CarmaxBrowser import CarmaxBrowser
from CarmaxAPI import CarmaxAPI
import numpy as np
```

```python
browser = CarmaxBrowser()
api = CarmaxAPI(browser)
records = api.search_model_headless_all(search='/cars?search=camry', zip_code=15206)
browser.close_browser()
```

```python
import pandas as pd

interested_col = ['year', 'make', 'model', 'trim', 'msrp', 'mileage', 'basePrice', 'isNew', 'highlights']

df = pd.DataFrame(records, columns=[i for i in interested_col])
df
```

Buy $y_{car}$ year car at 2019, with mileage $m_{in}$, in price $p_{in}$.

Sell at 2020, with price $p_{out}$ with mileage $m_{out}$ with mileage ${m_{out}}$.

$\Delta{year} = 2019 - \text{car['year']} $

$\Delta{mileage} = m_{out} - m_{in} $

```python
import sklearn as skl
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
x_scaled = scaler.fit_transform([i for i in zip(df['year'], df['mileage'])])
y_ = df['basePrice']
```

```python
model = LinearRegression().fit(x_scaled, y_)
model.score(x_scaled, y_)
```

```python
model.predict( scaler.transform( np.array([[2016, 70000]]) ))
```
